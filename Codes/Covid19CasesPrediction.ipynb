### 1.Confirmed Cases

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
from fbprophet import Prophet
```

```python
df = pd.read_csv(r"C:\Users\sonia\OneDrive\Desktop\MlProject\dataframe\covid_19_data.csv")
```

```python
df.head()
```

```python
df = df.sort_values('ObservationDate')
```

```python
#plot date and confirmed
plt.figure(figsize=(10,10))
plt.plot(df['ObservationDate'],df['Confirmed'])
```

```python
#plot distribution of the confirmed 
plt.figure(figsize=(10,6))
sns.distplot(df['Confirmed'],color='b')
```

```python
sns.set(font_scale=0.7)
plt.figure(figsize=[100,50])
sns.countplot(x='Country/Region',data=df[['Confirmed','Country/Region']])
plt.xticks(rotation=45)
```

```python
prophet_df = df[['ObservationDate','Confirmed']]
```

```python
prophet_df.head()
```

```python
prophet_df = prophet_df.rename(columns={'ObservationDate':'ds','Confirmed':'y'})
```

```python
m = Prophet()
m.fit(prophet_df)
```

```python
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast,xlabel='Date',ylabel='ConfirmedCase')
```

```python
figure2 = m.plot_components(forecast)
```
